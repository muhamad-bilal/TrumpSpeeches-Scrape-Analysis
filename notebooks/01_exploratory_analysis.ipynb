# Trump Speech Analysis - Exploratory Data Analysis

This notebook provides initial exploration of the cleaned speech data.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11

%matplotlib inline
```

## Load Cleaned Data


```python
# Find the most recent cleaned data file
from pathlib import Path


data_dir = Path('../data/cleaned')
csv_files = list[Path](data_dir.glob('speeches_cleaned_*.csv'))

if csv_files:
    # Sort by modification time and get the latest
    latest_file = max(csv_files, key=lambda p: p.stat().st_mtime)
    print(f"Loading: {latest_file.name}")
    df = pd.read_csv(latest_file)
    print(f"Loaded {len(df)} speeches")
else:
    print("No cleaned data found. Please run the data cleaning script first.")
```

## Basic Statistics


```python
# Display basic info
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst few rows:")
df.head()
```

```python
# Word count statistics
print("Word Count Statistics:")
print(df['word_count'].describe())

# Plot word count distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['word_count'], bins=30, edgecolor='black')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.title('Distribution of Speech Lengths')

plt.subplot(1, 2, 2)
plt.boxplot(df['word_count'])
plt.ylabel('Word Count')
plt.title('Speech Length Box Plot')

plt.tight_layout()
plt.show()
```
