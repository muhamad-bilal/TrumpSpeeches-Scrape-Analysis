# Rhetorical Analysis - Trump Speeches

Analysis of rhetorical devices: anaphora, repetition, contrast, alliteration.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_style('whitegrid')
%matplotlib inline

# Load features
data_dir = Path('../data/transformed')
csv_files = list(data_dir.glob('speeches_features_complete_*.csv'))
if csv_files:
    df = pd.read_csv(max(csv_files, key=lambda p: p.stat().st_mtime))
    print(f"Loaded {len(df)} speeches")
else:
    print("No data found")
```

## Anaphora Patterns


```python
# Anaphora analysis
anaphora_cols = [c for c in df.columns if c.startswith('anaphora_')]

if anaphora_cols:
    anaphora_data = df[anaphora_cols].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    anaphora_data.plot(kind='bar')
    plt.title('Anaphora Pattern Usage Across All Speeches')
    plt.xlabel('Pattern')
    plt.ylabel('Total Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
